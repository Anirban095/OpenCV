```python
image = cv2.imread('./images/flowers.jpeg')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def imageShow(title ='', image = None):
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
imageShow('First Image', image)
```
